```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Upload from CSVs
savage_path = "SavageGradesDataCSV2.csv"
lawrence_path = "lawrence_ela.csv"

#Read data and results
savage_df = pd.read_csv(savage_path)
lawrence_df = pd.read_csv(lawrence_path)

savage_df.head()
```

```python
#Savage Data Types
savage_df.dtypes
```

```python
savage_df["Percent"] = savage_df["Score"]*100
savage_df["Percent"] = savage_df["Percent"].astype("int64")

savage_df.head()
```

```python
#Write as letter grade
letter_grades=[]

for row in savage_df["Percent"]:
    if row >= 90:
        letter_grades.append("A")
    elif row >=80:
        letter_grades.append("B")
    elif row >=70:
        letter_grades.append("C")
    elif row >=60:
        letter_grades.append("D")
    elif row <60:
        letter_grades.append("F")

#Create new column
savage_df["letter_grades"]=letter_grades
savage_df.head()
```

```python
#Savage

#Hours
savage_2_df = savage_df.loc[savage_df["Hour"]==2]
savage_3_df = savage_df.loc[savage_df["Hour"]==3]
savage_5_df = savage_df.loc[savage_df["Hour"]==5]
savage_6_df = savage_df.loc[savage_df["Hour"]==6]
savage_7_df = savage_df.loc[savage_df["Hour"]==7]

#Mean
savage_mean = savage_df["Percent"].mean()
savage_2_mean = savage_2_df["Percent"].mean()
savage_3_mean = savage_3_df["Percent"].mean()
savage_5_mean = savage_5_df["Percent"].mean()
savage_6_mean = savage_6_df["Percent"].mean()
savage_7_mean = savage_7_df["Percent"].mean()


#Median
savage_median = savage_df["Percent"].median()
savage_2_median = savage_2_df["Percent"].median()
savage_3_median = savage_3_df["Percent"].median()
savage_5_median = savage_5_df["Percent"].median()
savage_6_median = savage_6_df["Percent"].median()
savage_7_median = savage_7_df["Percent"].median()


#Max
savage_max = savage_df["Percent"].max()
savage_2_max = savage_2_df["Percent"].max()
savage_3_max = savage_3_df["Percent"].max()
savage_5_max = savage_5_df["Percent"].max()
savage_6_max = savage_6_df["Percent"].max()
savage_7_max = savage_7_df["Percent"].max()

# #Min
savage_min = savage_df["Percent"].min()
savage_2_min = savage_2_df["Percent"].min()
savage_3_min = savage_3_df["Percent"].min()
savage_5_min = savage_5_df["Percent"].min()
savage_6_min = savage_6_df["Percent"].min()
savage_7_min = savage_7_df["Percent"].min()

# #Range
savage_range = savage_max - savage_min
savage_2_range = savage_2_max - savage_2_min
savage_3_range = savage_3_max - savage_3_min
savage_5_range = savage_5_max - savage_5_min
savage_6_range = savage_6_max - savage_6_min
savage_7_range = savage_7_max - savage_7_min

index = ["Mean", "Median", "Range", "Min", "Max"]
savage_data_df=pd.DataFrame({
                           "All Hours": [savage_mean, savage_median, savage_range, savage_min, savage_max],
                            "Hour 2": [savage_2_mean, savage_2_median, savage_2_range, savage_2_min, savage_2_max],
                            "Hour 3": [savage_3_mean, savage_3_median, savage_3_range, savage_3_min, savage_3_max],
                            "Hour 5": [savage_5_mean, savage_5_median, savage_5_range, savage_5_min, savage_5_max],
                            "Hour 6": [savage_6_mean, savage_6_median, savage_6_range, savage_6_min, savage_6_max],
                            "Hour 7": [savage_7_mean, savage_7_median, savage_7_range, savage_7_min, savage_7_max]}, index = index)

# #Change to int
savage_data_df["All Hours"] = savage_data_df["All Hours"].astype("int64")
savage_data_df["Hour 2"] = savage_data_df["Hour 2"].astype("int64")
savage_data_df["Hour 3"] = savage_data_df["Hour 3"].astype("int64")
savage_data_df["Hour 5"] = savage_data_df["Hour 5"].astype("int64")
savage_data_df["Hour 6"] = savage_data_df["Hour 6"].astype("int64")
savage_data_df["Hour 7"] = savage_data_df["Hour 7"].astype("int64")

savage_data_df

transpose_savage_df = savage_data_df.transpose()

transpose_savage_df
```

```python
#Lawrence

#Convert Data Types
lawrence_df["Percent"] = lawrence_df["Score"]*100
lawrence_df["Percent"] = lawrence_df["Percent"].astype("int64")

#Assign Letter Grades
letter_grades=[]

for row in lawrence_df["Percent"]:
    if row >= 90:
        letter_grades.append("A")
    elif row >=80:
        letter_grades.append("B")
    elif row >=70:
        letter_grades.append("C")
    elif row >=60:
        letter_grades.append("D")
    elif row <60:
        letter_grades.append("F")

#Create new column
lawrence_df["letter_grades"]=letter_grades

lawrence_df.head()
```

```python
#Lawrence Count and Sort
lawrence_counts_df = (lawrence_df["letter_grades"].value_counts())

lawrence_counts_df = pd.DataFrame(lawrence_counts_df)
lawrence_counts_df = lawrence_counts_df.sort_values(by=["letter_grades"])
lawrence_counts_df.head()
```

```python
#Lawrence
#Mean
lawrence_mean = lawrence_df["Percent"].mean()
#Median
lawrence_median = lawrence_df["Percent"].median()
#Max
lawrence_max = lawrence_df["Percent"].max()
# #Min
lawrence_min = lawrence_df["Percent"].min()
#Range
lawrence_range = lawrence_max - lawrence_min
index = ["Mean", "Median", "Range", "Min", "Max"]
lawrence_data_df=pd.DataFrame({
                           "All Hours": [lawrence_mean, lawrence_median, lawrence_range, lawrence_min, lawrence_max]}, index = index)
#Change to int
lawrence_data_df["All Hours"] = lawrence_data_df["All Hours"].astype("int64")
lawrence_data_df
transpose_lawrence_df = lawrence_data_df.transpose()

transpose_lawrence_df
```

```python
#Merge Teachers
merged_teachers_df = pd.merge(transpose_savage_df, transpose_lawrence_df, how="outer")
merged_teachers_df = merged_teachers_df.iloc[[0,5]]
merged_teachers_df = merged_teachers_df.rename(index={0:"Math", 5:"ELA"})
merged_teachers_df = merged_teachers_df.transpose()
merged_teachers_df
```

```python
#Savage Letter Grade Counts
#Counts by grade
savage_grade_2_df = (savage_2_df["letter_grades"].value_counts())
savage_grade_3_df = (savage_3_df["letter_grades"].value_counts())
savage_grade_5_df = (savage_5_df["letter_grades"].value_counts())
savage_grade_6_df = (savage_6_df["letter_grades"].value_counts())
savage_grade_7_df = (savage_7_df["letter_grades"].value_counts())

#DataFrame
savage_grade_2_df = pd.DataFrame(savage_grade_2_df)
savage_grade_3_df = pd.DataFrame(savage_grade_3_df)
savage_grade_5_df = pd.DataFrame(savage_grade_5_df)
savage_grade_6_df = pd.DataFrame(savage_grade_6_df)
savage_grade_7_df = pd.DataFrame(savage_grade_7_df)


#Merge and Rename 1
merged_df = pd.merge(savage_grade_2_df, savage_grade_3_df,how="outer", on = "letter_grades")

merged_df = merged_df.rename(columns = {"count_x":"Hour 2", "count_y":"Hour 3"})

#Merge and Rename 2
merged_df = pd.merge(merged_df, savage_grade_5_df, how="outer", on="letter_grades")
merged_df = merged_df.rename(columns = {"count":"Hour 5"})

#Merge and Rename 3
merged_df = pd.merge(merged_df, savage_grade_6_df, how="outer", on="letter_grades")
merged_df = merged_df.rename(columns={"count":"Hour 6"})

#Merge and Rename 4
merged_df = pd.merge(merged_df, savage_grade_7_df, how="outer", on="letter_grades")
merged_df = merged_df.rename(columns={"count":"Hour 7"})
merged_df = merged_df.fillna(0)

#Astype
merged_df["Hour 2"] = merged_df["Hour 2"].astype("int64")
merged_df["Hour 7"] = merged_df["Hour 7"].astype("int64")

#Sort
savage_grade_2_df = savage_grade_2_df.sort_values(by=["letter_grades"])
merged_df = merged_df.sort_values(by=["letter_grades"])
#merged_df.head()

transpose_df = merged_df.transpose()

#transpose_df.head()
merged_df.head()
```

```python
#Savage Piechart All
labels = all_df.index
sizes = all_df["All"]
colors = ["paleturquoise", "lightgreen", "thistle", "palevioletred", "yellow"]

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format
    
plt.title("Savage Scores")
plt.pie(sizes, labels=labels, colors=colors, autopct=autopct_format(sizes), startangle=120)
plt.savefig('SavageAllPie.png')
plt.show()

```
