```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Upload from CSVs
math_path = "RakonickGrades.csv"
lawrence_path = "lawrence_ela.csv"

#Read data and results
math_df = pd.read_csv(math_path)
lawrence_df = pd.read_csv(lawrence_path)

math_df.head()
```

```python
#Savage Data Types
math_df.dtypes
```

```python
math_df["Percent"] = math_df["Score"]*100
math_df["Percent"] = math_df["Percent"].astype("int64")

math_df.head()
```

```python
#Write as letter grade
letter_grades=[]

for row in math_df["Percent"]:
    if row >= 90:
        letter_grades.append("A")
    elif row >=80:
        letter_grades.append("B")
    elif row >=70:
        letter_grades.append("C")
    elif row >=60:
        letter_grades.append("D")
    elif row <60:
        letter_grades.append("F")

#Create new column
math_df["letter_grades"]=letter_grades
math_df.head()
```

```python
#Savage

#Hours
math_1_df = math_df.loc[math_df["Hour"]==1]
math_2_df = math_df.loc[math_df["Hour"]==2]
math_6_df = math_df.loc[math_df["Hour"]==6]
math_7_df = math_df.loc[math_df["Hour"]==7]


#Mean
math_mean = math_df["Percent"].mean()
math_1_mean = math_1_df["Percent"].mean()
math_2_mean = math_2_df["Percent"].mean()
math_6_mean = math_6_df["Percent"].mean()
math_7_mean = math_7_df["Percent"].mean()


#Median
math_median = math_df["Percent"].median()
math_1_median = math_1_df["Percent"].median()
math_2_median = math_2_df["Percent"].median()
math_6_median = math_6_df["Percent"].median()
math_7_median = math_7_df["Percent"].median()


#Max
math_max = math_df["Percent"].max()
math_1_max = math_1_df["Percent"].max()
math_2_max = math_2_df["Percent"].max()
math_6_max = math_6_df["Percent"].max()
math_7_max = math_7_df["Percent"].max()


# #Min
math_min = math_df["Percent"].min()
math_1_min = math_1_df["Percent"].min()
math_2_min = math_2_df["Percent"].min()
math_6_min = math_6_df["Percent"].min()
math_7_min = math_7_df["Percent"].min()


# #Range
math_range = math_max - math_min
math_1_range = math_1_max - math_1_min
math_2_range = math_2_max - math_2_min
math_6_range = math_6_max - math_6_min
math_7_range = math_7_max - math_7_min


index = ["Mean", "Median", "Range", "Min", "Max"]
math_data_df=pd.DataFrame({
                           "All Hours": [math_mean, math_median, math_range, math_min, math_max],
                            "Hour 1": [math_1_mean, math_1_median, math_1_range, math_1_min, math_1_max],
                            "Hour 2": [math_2_mean, math_2_median, math_2_range, math_2_min, math_2_max],
                            "Hour 6": [math_6_mean, math_6_median, math_6_range, math_6_min, math_6_max],
                            "Hour 7": [math_7_mean, math_7_median, math_7_range, math_7_min, math_7_max]}, index = index)

# #Change to int
math_data_df["All Hours"] = math_data_df["All Hours"].astype("int64")
math_data_df["Hour 1"] = math_data_df["Hour 1"].astype("int64")
math_data_df["Hour 2"] = math_data_df["Hour 2"].astype("int64")
math_data_df["Hour 6"] = math_data_df["Hour 6"].astype("int64")
math_data_df["Hour 7"] = math_data_df["Hour 7"].astype("int64")

math_data_df

transpose_math_df = math_data_df.transpose()

math_data_df.head()
```

```python
#Lawrence

#Convert Data Types
lawrence_df["Percent"] = lawrence_df["Score"]*100
lawrence_df["Percent"] = lawrence_df["Percent"].astype("int64")

#Assign Letter Grades
letter_grades=[]

for row in lawrence_df["Percent"]:
    if row >= 90:
        letter_grades.append("A")
    elif row >=80:
        letter_grades.append("B")
    elif row >=70:
        letter_grades.append("C")
    elif row >=60:
        letter_grades.append("D")
    elif row <60:
        letter_grades.append("F")

#Create new column
lawrence_df["letter_grades"]=letter_grades

lawrence_df.head()
```

```python
#Lawrence Count and Sort
lawrence_counts_df = (lawrence_df["letter_grades"].value_counts())

lawrence_counts_df = pd.DataFrame(lawrence_counts_df)
lawrence_counts_df = lawrence_counts_df.sort_values(by=["letter_grades"])
lawrence_counts_df.head()
```

```python
#Lawrence
#Mean
lawrence_mean = lawrence_df["Percent"].mean()
#Median
lawrence_median = lawrence_df["Percent"].median()
#Max
lawrence_max = lawrence_df["Percent"].max()
# #Min
lawrence_min = lawrence_df["Percent"].min()
#Range
lawrence_range = lawrence_max - lawrence_min
index = ["Mean", "Median", "Range", "Min", "Max"]
lawrence_data_df=pd.DataFrame({
                           "All Hours": [lawrence_mean, lawrence_median, lawrence_range, lawrence_min, lawrence_max]}, index = index)
#Change to int
lawrence_data_df["All Hours"] = lawrence_data_df["All Hours"].astype("int64")
lawrence_data_df
transpose_lawrence_df = lawrence_data_df.transpose()

transpose_lawrence_df
```

```python
#math Letter Grade Counts
#Counts by grade
math_grade_1_df = (math_1_df["letter_grades"].value_counts())
math_grade_2_df = (math_2_df["letter_grades"].value_counts())
math_grade_6_df = (math_6_df["letter_grades"].value_counts())
math_grade_7_df = (math_7_df["letter_grades"].value_counts())


#DataFrame
math_grade_1_df = pd.DataFrame(math_grade_1_df)
math_grade_2_df = pd.DataFrame(math_grade_2_df)
math_grade_6_df = pd.DataFrame(math_grade_6_df)
math_grade_7_df = pd.DataFrame(math_grade_7_df)


#Merge and Rename 1
merged_df = pd.merge(math_grade_1_df, math_grade_2_df,how="outer", on = "letter_grades")

merged_df = merged_df.rename(columns = {"count_x":"Hour 1", "count_y":"Hour 2"})

#Merge and Rename 2
merged_df = pd.merge(merged_df, math_grade_6_df, how="outer", on="letter_grades")
merged_df = merged_df.rename(columns = {"count":"Hour 6"})

#Merge and Rename 3
merged_df = pd.merge(merged_df, math_grade_7_df, how="outer", on="letter_grades")
merged_df = merged_df.rename(columns = {"count":"Hour 7"})

#Sort
merged_df = merged_df.sort_values(by=["letter_grades"])


transpose_df = merged_df.transpose()

merged_df = merged_df.fillna(0)
merged_df["Hour 2"] = merged_df["Hour 2"].astype("int64")
merged_df["Hour 7"] = merged_df["Hour 7"].astype("int64")
merged_df
```

```python
#math Bar Graph
#Group plots by grade
multi_plot = transpose_df.plot(kind="bar", rot=0)

plt.legend()
plt.grid(axis = "y")
plt.xlabel("Hour")
plt.ylabel("Number of Students")
plt.title("Grades by Hour")
plt.savefig('math.png')
plt.show()


```

```python
#math Piechart All
labels = all_df.index
sizes = all_df["All"]
colors = ["paleturquoise", "lightgreen", "thistle", "palevioletred", "yellow"]

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format
    
plt.title("Rakonick Scores")
plt.pie(sizes, labels=labels, colors=colors, autopct=autopct_format(sizes), startangle=120)
plt.savefig('RakonickAllPie.png')
plt.show()

```
